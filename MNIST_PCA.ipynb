```python
# import module
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import pandas as pd
```

```python
# lode csv file
data = pd.read_csv('./MNIST_100.csv')
data
```

```python
# data 전처리
X = data.drop('label', axis=1)
y = data.iloc[:, 0]
```

```python
# PCA 그래프 처리
pca = PCA(n_components=2)
pca.fit(X)
PCAX = pca.transform(X)

plt.title("MNIST PCA")
plt.xlabel("X")
plt.ylabel("Y")
for i in range(0,901,100) : 
    plt.scatter(PCAX[i:i+100, 0], PCAX[i:i+100, 1])
plt.show()
```

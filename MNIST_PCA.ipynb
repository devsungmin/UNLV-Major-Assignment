```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import pandas as pd
```

```python
data = pd.read_csv('./MNIST_100.csv')
```

```python
data
```

```python
y = data.iloc[:, 0]
X = data.drop('label', axis=1)
```

```python
pca = PCA(n_components=2)
pca.fit(X)
PCAX = pca.transform(X)


for i in range(0,901,100) : 
    plt.scatter(PCAX[i:i+100, 0], PCAX[i:i+100, 1])
plt.show()
```
